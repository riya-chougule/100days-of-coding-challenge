### Description : In this notebook, I have implemented EDA on a dataset from kaggle (IPL Complete Dataset (2008-2020)). For this EDA, I used python libraries like pandas, numpy, matplotlib, seaborn, missingno.
### The dataset includes two csv files :
#### i) IPL Matches 2008-2020.csv
#### ii) IPL Ball-by-Ball 2008-2020.csv

#### Link to the dataset : [IPL Complete dataset (2008-2020)](https://www.kaggle.com/patrickb1912/ipl-complete-dataset-20082020)

### Date : 27-09-2021

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import missingno
%matplotlib inline
```

```python
df1 = pd.read_csv('IPL Matches 2008-2020.csv')
```

```python
df1.info()
```

```python
df1.head()
```

```python
df2 = pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
```

```python
df2.info()
```

```python
df2.head()
```

```python
df1['toss_decision'].value_counts().plot(kind='bar', title='Decision of toss', figsize=(9,7))
```

```python
df1['city'].value_counts().plot(kind='bar', title='City', figsize=(9,7))
```

```python
df2['batsman_runs'].value_counts().plot(kind='bar', title="Batsman's run info", figsize=(9,7))
```

```python
sns.histplot(df1['result_margin'],stat="density", linewidth=2)
```

```python
sns.histplot(df2['batsman_runs'],stat="density", linewidth=2)
```

```python
sns.relplot(x='team1', y='team2', hue='result', data=df1,height=7, aspect=3.4)
```

```python
sns.relplot(x='over', y='ball', hue='is_wicket', data=df2)
```

```python
df2.groupby('over')[['batsman_runs', 'total_runs', 'ball']].mean().T.plot(figsize=(12,8))
```

```python
df1.groupby('over')[['batsman_runs', 'total_runs', 'ball']].mean().T.plot(figsize=(12,8))
```

```python
corr = df1.corr()
sns.heatmap(corr, annot=True, square=True)
plt.yticks(rotation=0)
plt.show()
```

```python
corr = df2.corr()
sns.heatmap(corr, annot=True, square=True)
plt.yticks(rotation=0)
plt.show()
```
